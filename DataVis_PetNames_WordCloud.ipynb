```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
```

```python
train_df = pd.read_excel('/Users/carinaland/Documents/Thesis/05 Data/train_df.xlsx')

text_cat = ','.join(str(v) for v in train_df.loc[train_df['Type'] == 2].Name)
text_dog = ','.join(str(v) for v in train_df.loc[train_df['Type'] == 1].Name)

text_cat = text_cat.replace("nan,", "")
text_dog = text_dog.replace("nan,", "")
```

```python
def word_cloud(text):
    # Create a word cloud image
    wc = WordCloud(width=1500, height=1000, background_color="white", max_words=300,
                   contour_width=3, contour_color='firebrick')

    # Generate a wordcloud
    wc.generate(text)

    # show
    plt.figure(figsize=[15, 8])
    plt.imshow(wc, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python
word_cloud(text_cat)
```

```python
word_cloud(text_dog)
```
